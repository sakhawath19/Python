```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
# it creates one figure and 4 axis or 4 plot
fig, ax = plt.subplots(nrows=2, ncols=2)
print(ax)
ax[0][0].set_title('first one')
ax[0][1].set_title('first row second one')
ax[1][0].set_title('second row first one')
ax[1][1].set_title('second row second plot')
```

```python
# unpacking plot with different figures

fig1, ax1 = plt.subplots()
ax1.set_title('This is plotted in first figure')

fig2, (ax2, ax3) = plt.subplots(nrows=2, ncols=1)

ax2.set_title('This is plotted in second figure')
ax2.set_ylabel('Y grid')
ax3.set_xlabel('X grid')
```

```python
# object oriented style plotting

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0,0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0,1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1,0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1,1]')

plt.tight_layout()
plt.show()

# fig1.savefig('fig1.png')
# fig2.savefig('fig2.png')
```

```python
# normal style plotting

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')
```

```python
# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')
```

```python
# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')
```

```python
# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)
```
