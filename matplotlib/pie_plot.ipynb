```python
# pie chart does not look good with more data
# 5 data point is a good number to show

### color code
### edge color
### hexcode of the color
### emphasize specific part
### change the orientation
### change 
import matplotlib.pyplot as plt 

plt.style.use('fivethirtyeight')
# plt.style.available
# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

```python
### default color palette in fivethirtyeight 
slices = [120, 80, 30, 20]
labels = ['sixty', 'forty', 'extra1', 'extra2']

# wedgeprops added black color in edge 
plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
### edgecolor 
### defined color palette
slices = [120, 80, 30, 20]
labels = ['sixty', 'forty', 'extra1', 'extra2']
colors = ['blue', 'red', 'yellow', 'green']

# colors does not look good in the plot. Better to use hex color
# wedgeprops added black color in edge 
plt.pie(slices, labels=labels, colors=colors,
       wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
### Hexcode
slices = [120, 80, 30, 20]
labels = ['sixty', 'forty', 'extra1', 'extra2']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

# Hexcode improved the quality 
plt.pie(slices, labels=labels, colors=colors,
       wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
### explode to emphasize 
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

# explode used to emphasize specific part of the chart
# 0.1 means only Python will be exploded upto 10% of its radius
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, colors=colors, explode=explode,  
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
### shadow
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

explode = [0, 0, 0, 0.1, 0]

# shadow will create 3D effect 
plt.pie(slices, labels=labels, colors=colors, explode=explode, shadow=True,
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
### startangle to fix the orientation of the plot

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

explode = [0, 0, 0, 0.1, 0]

# startangle 90 degree will create exactly vertical line with first label
# and then will place other value counter clockwise
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90,
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
### autopct to represent percentage of the data  

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

explode = [0, 0, 0, 0.1, 0]

# shadow will create 3D effect 
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
