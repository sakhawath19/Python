```python
import pandas as pd
%matplotlib inline 
from matplotlib import pyplot as plt
import os 

plt.style.use('seaborn')
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]

y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.figure(figsize=(5, 4)) 
plt.scatter(x, y)

plt.tight_layout()

plt.show()
```

```python
### simple scatter plot
# x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
# y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

# The x array represents the age of each car.
# The y array represents the speed of each car.
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]

y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.figure(figsize=(5, 4))
plt.scatter(x, y)

plt.tight_layout()

plt.show()
```

```python
### color, marker, size, edge color, linewidth, color transparency 
# x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
# y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.figure(figsize=(5, 4))
# color transparency changes with alpha value
plt.scatter(x, y, s=50, c='green', edgecolors='black', marker='o', linewidths=1, alpha=0.5)

plt.tight_layout()

plt.show()
```

```python
### color shade and color map 

# x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
# y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

colors = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.figure(figsize=(5, 4))

# cmap tells the specific color
# c chooses the different shade 
plt.scatter(x, y, s=100, c=colors, cmap='Greens', 
            edgecolors='black', marker='o', linewidths=1, alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()

plt.show()
```

```python
### scatter plot with real data
### outlier skewing the plot, in that case better to use log scale, see next plot

data = pd.read_csv(os.getcwd() + './data/scatter_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolors='black', linewidths=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
### log scale gives better view 

data = pd.read_csv(os.getcwd() + './data/scatter_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolors='black', linewidths=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')  

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
### log scale gives better view 

data = pd.read_csv(os.getcwd() + './data/scatter_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer', 
            edgecolors='black', linewidths=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')  

cbar = plt.colorbar()
cbar.set_label('like dislike ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
