```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# by default  kind='strip'
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(y='day', x='total_bill', data=tips)
```

```python
# remove deviation using jitter 
sns.catplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
# swarmplot avoid overlapping data 
sns.catplot(x='day', y='total_bill', kind='swarm', hue='sex', data=tips)
```

```python
sns.catplot(x='smoker', y='tip', data=tips, order=['No', 'Yes'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', data=tips)
```

```python
# dodge=False made the graph overlap each other
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', data=tips, dodge=False)
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
# kind='boxen'

sns.catplot(x='color', y='price', kind='boxen', data=diamonds)
```

```python
sns.catplot(x='color', y='price', kind='boxen', data=diamonds.sort_values('color'))
```

```python
sns.catplot(x='total_bill', y='day', hue='time', kind='violin', data=tips)
```

```python
# splitting to multiple violin using split=True
# violin plot is a combination of kdeplot and boxplot
sns.catplot(x='total_bill', y='day', hue='sex', kind='violin', split=True, data=tips, inner='sticks')
```

```python
# plotted swarmplot inside the violit plot using the axis of violin
g = sns.catplot(x='day', y='total_bill', kind='violin', data=tips, inner=None)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips, ax=g.ax)
```

```python
# observing two categorical data 
sns.catplot(x='sex', y='survived', hue='class', kind='bar', data=titanic)
```

```python
# change the shades of the color 
sns.catplot(x='deck', kind='count', palette='ch:0.55', data=titanic)
```

```python
sns.catplot(x='deck', kind='count', palette='ch:0.55', hue='class', data=titanic)
```

```python
# plot with annotation 
plt.figure(figsize=(6, 8))
splot = sns.barplot(data=tips, x = 'sex', y = 'total_bill', ci = None)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```
