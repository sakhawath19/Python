```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.tail()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
dir(sns.FacetGrid)
```

```python
dir(sns.axisgrid)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker') # categorical plotting 
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='time') 
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size') # qualitative plotting
```

```python
# hue changes the color
# style changes the object
# size changes the size of the object 

# here default is scatter plot, plot can be changed using kind argument

# available seaborn color palette deep, muted, bright, pastel, dark, colorblind
# we customize a sequential palette using the string interface to cubehelix_palette():
# palette="ch:r=-.5,l=.75"
# sns.palplot(sns.color_palette("RdBu", n_colors=7))

# sns.relplot(x='total_bill', y='tip', data=tips, palette='muted')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette="ch:r=-.5,l=.75") 
```

```python
# size of the object will be normalized in between 15 to 200
sns.relplot(x='total_bill', y='tip', data=tips, size='size', sizes=(15, 200)) 
```

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time= np.arange(500), value= randn(500).cumsum()))
```

```python
sns.relplot(x='time', y='value', data=df, kind='line')
```

```python
df = pd.DataFrame(randn(500, 2).cumsum(axis=0), columns=['time', 'value'])
```

```python
df.head()
```

```python
sns.relplot(x='time', y='value', data=df, kind='line', sort=False)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
# sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci=False) 
# by default showing the confidence interval(ci) when one timestamp has multiple value 

sns.relplot(x='timepoint', y='signal', kind='line', data=fmri)
```

```python
# standard deviation is defined as confidence interval
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci='sd')
```

```python
# plotting original data 
sns.relplot(x='timepoint', y='signal', estimator=None, kind='line', data=fmri)
```

```python
# plotting categorical data with confidence interval

sns.relplot(x='timepoint', y='signal', hue='event', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', hue='region', style='event', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', hue='region', style='event', kind='line', data=fmri, markers = True, dashes=False)
```

```python
sns.relplot(x='timepoint', y='signal', hue='event', style='event', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', hue='region', units='subject', estimator=None, kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', hue='region', units='subject', estimator=None, kind='line', data=fmri.query("event=='stim'"))
```
